```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open("1.8um_Data/pmos_data/pmos_2b.txt", 'r') as file:
    lines = [line.rstrip() for line in file][6:]

    voltage = []
    current = []

    for line in lines:
        b, a = line.split("      ")
        voltage.append(float(b))
        current.append(float(a))

    plt.plot(voltage, np.abs(current), label="New Channel Doping: 1.395e16")

    plt.yscale('log')

    plt.vlines(0, 1e-13, 1e-5, color='black')

    plt.scatter(0, 1.01e-9, label="New Off State Current: 1.01e-9", zorder=3, color='blue')


with open("1.8um_Data/pmos_data/pmos_1a_high.txt", 'r') as file:
    lines = [line.rstrip() for line in file][6:]

    voltage = []
    current = []

    for line in lines:
        b, a = line.split("      ")
        voltage.append(float(b))
        current.append(float(a))

    plt.plot(voltage, np.abs(current), label="Original Channel Doping: 3e16")

    plt.yscale('log')

    plt.scatter(0, 1.15e-11, label="Original Off State Current: 1.15e-11", zorder=3, color='red')


plt.ylabel("Absolute Current Density (|A|/um)")
plt.xlabel("Gate Voltage Vg")
plt.title("Pmos Off Current Comparison (2b)")
plt.legend(loc=3)

plt.show()
```
